# Clean Data

**Remove the unnecessary columns from the dataset**

```python
extra_cols=['Unnamed: 0','Url_spotify','Uri','Url_youtube', 'Title']
```

```python
df.drop(extra_cols,axis=1,inplace=True)
```

Here we have removed 'Unnamed: 0','Title','Url_spotify','Uri' and 'Url_youtube' columns from the dataset beacsuse they are not in need for analysis

```python
#viewing the number of unique value of the each column
df.nunique()
```

we can see here ,we have the unique value of each columns as we have 2079 artist,17841 track ,3 album_type ,11232 album and other columns uniquevalues

```python
#checking the number of null values in each column
df.isna().sum()
```

now we can see here the number of null values in each column we have more null values in
channel:470,Views:470,Comments:470,Licensed:470 andDescription:876 etc

```python
#removing all the null values from the columns
df.dropna(inplace=True)
```

```python
#now again rechecking the null value (if there any missing)
df.isnull().sum()
```

```python
df.shape
```

after removing the unnecessary columns and null values from columns ,now we have 19170 rows and 23 columns

**Changing the data type of the columns as per need**

```python
df['Views']=df['Views'].astype('int')
df['Likes']=df['Likes'].astype('int')
df['Comments']=df['Comments'].astype('int')
```

**Determining the top 10 artists on the Spotify platform whose songs have more views**

```python
views=df.groupby('Artist')['Views'].sum().reset_index().sort_values('Views',ascending=False).head(10)
views
```

```python
views['Views'].apply('{:e}'.format,inplace=True)
views
```

```python
plt.rcParams['figure.figsize'] = (8,6)
plt.bar(views['Artist'],views['Views'],color='green')
plt.xlabel('Artist')
plt.ylabel('Views')
plt.title('Top 10 Artist have more views')
plt.xticks(rotation=70)
plt.show()
```

we can see here Bruno mars was the most viewed artist with 10.23 billion views.Macklemore & Ryan Lewis and Macklemore & Ryan Lewis are the secondand third artist with 10.12 billion views and 10 billion respectively.These figures highlight the popularity of these artists and their music, as well as the significantrole of online video platforms in promoting and distributing music.

**Display HeatMap**
